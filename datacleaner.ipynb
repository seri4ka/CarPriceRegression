```python
import numpy as np
import pandas as pd
import sklearn as sl
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
```

```python
data = pd.read_csv('carsutf.csv', sep=',')
```

```python
data
```

```python
df['Mark'] = data['Марка']
df['Model'] = data['Модель']
df['Year'] = data['Год выпуска']
df['Oil'] = data['Двигатель'].str.split(', ', expand=True)[0]
df['Volume'] = data['Двигатель'].str.split(', ', expand=True)[1].str.split(' ', expand=True)[0]
df['Power'] = data['Мощность'].str.split(expand=True)[0]
df['GSB'] = data['КПП']
df['Drive'] = data['Привод']
df['Color'] = data['Цвет']
df['Kilometers'] = data['Пробег'].str.split().str.join('')
df['Wheel'] = data['Руль']
df['Generation'] = data['Поколение']
df['Complectation'] = data['Комплектация']
df['Price'] = data['Цена']
```

```python
df
```

```python
df.to_csv('ClearCarInform.csv')
```
